```python
# Import dependencies
import numpy as np
import pandas as pd
```

```python
# Read the data
df = pd.read_csv("sample_load_weather_data.csv")
```

```python
df.head()
```

```python
# drop the unrelated columns
dropped_df = df.drop(["date", "wkday", "year"] , axis = 1, inplace=False)

dropped_df.tail(5)
```

```python
# Get dummies

df_dummies = pd.get_dummies(dropped_df)
df_dummies.head()
```

```python
# save the data
df_dummies.to_csv('dummies_data.csv')
```
